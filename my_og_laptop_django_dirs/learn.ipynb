some edits that i made to make a dafault project run i django

1. you ghave to create ur own urls.py 
2. you have to create ur own views.py
3. you have to create ur own templates folder in root folder
4. you have to create ur own static folder in root folder

- edit the *settings.py* file
1. in the templates section add the following code<br>
   ` DIRS': ['templates'], or 'DIRS': [os.path.join(BASE_DIR, 'templates')],`
2. in the static section add the following code <br>
>    ` STATIC_URL = '/static/'` <br>
>   ` STATICFILES_DIRS = [os.path.join(BASE_DIR, 'static')]`
3. in the installed apps section add the following code.  
   ` 'app.apps.AppConfig',`

- edit the **urls.py** file

use this to fix it 
- open the command palette (Ctrl+shift+P for win) and:
- type "Python select interpeter" --> "select interpreter".
- Choose "+ select interpreter path"
- Choose "Find... browse your files.."
- Go to your Django project folder, open your virtual enviroment folder .venv/scripts/...
- In that folder select the Python interpreter and that should be it.

## load html file to views.py
- in views.py file
> from django.shortcuts import render
>    return render(request, 'index.html')
- set the path to the templates folder in the settings.py file
- under the templates section add the following code
> 'DIRS': ['templates'], or 'DIRS': [os.path.join(BASE_DIR, 'templates')],
- create a new folder in the root folder and name it templates
- create a new html file in the templates folder and name it index.html
- add some html code to the index.html file



## load css to html file
-in settings.py file
> STATIC_URL = 'static/'
> STATICFILES_DIRS = [
    BASE_DIR / "static",
]
- in the html file
> {% load static %}  
`<link rel="stylesheet" type="text/css" href="{% static 'css/style.css' %}">`
